# Find the information for a specific Slot, Card, Patch, Reboot Cause reason.

This task helps in cases where a specific slot has got reported to have suddenly rebooted/crashed, leaving a trace about how the LPU/MPU/Slot crashed (for example, "Kernel Reboot," "User/Cold Reboot"). If it's a Chassis-based switch, when the input function is invoked, the user is prompted to input the specific Chassis and Slot of interest or if its slot then just mention slot you want, or you can look for any other specific information such as the Patch or Reboot Cause in the keyword prompt.

For example:

When asked, "What is the slot number?" the user might respond with> Chassis 1 Slot 1
When asked, "What is the keyword?" the user might respond with> Patch

Note: Replace the file path in the code with the location of your diagnostic information data, which should be in a text format. For example, replace "file = 'file path'" with "file = 'diag_pind_01032021.txt'" within the double-quote string."

```python
def grab_section(file, start_keyword, end_keyword):
    start = False
    section = []
    try:
        with open(file, 'r', encoding='utf-8') as f:
            for line in f:
                if start_keyword.lower().strip() in line.lower().strip():
                    start = True
                if start:
                    section.append(line)
                    if end_keyword.lower().strip() in line.lower().strip():
                        break
        return section
    except UnicodeDecodeError:
        with open(file, 'r', encoding='ansi') as f:
            for line in f:
                if start_keyword.lower().strip() in line.lower().strip():
                    start = True
                if start:
                    section.append(line)
                    if end_keyword.lower().strip() in line.lower().strip():
                        break
        return section

file = "icmp.txt"
start_keyword = input("What is chassis or/& slot number ? ")
end_keyword = input("What is keyword ? ")
section = grab_section(file, start_keyword, end_keyword)
print("\n" + ''.join(section))
```

# Query for MAC ADDRESS from the MAC & ARP table.

This task assist in cases where you would quickly want to find the existence of a particular mac-address output from the display mac-address & display arp all table.

When the cell prompts to type in "Enter the MAC address to search for:" < Key in the mac-address that you are looking for >

Note- Replace the file = 'file path' with "xxxx.txt" to display diagnostic-information file .

```python
def search_mac(file, mac_address):
    arp_results = []
    mac_results = []
    with open(file, "r", encoding='ansi') as f:
        in_arp_section = False
        in_mac_section = False
        for line in f:
            line = line.strip()
            if "display arp all" in line:
                in_arp_section = True
                continue
            elif "display mac-address" in line:
                in_arp_section = False
                in_mac_section = True
                continue
            elif not line or "-----" in line:
                in_mac_section = False
            elif in_arp_section:
                data = line.split()
                if len(data) >= 3 and data[1].lower() == mac_address.lower():
                    arp_result = {}
                    arp_result["IP address"] = data[0]
                    arp_result["MAC address"] = data[1]
                    arp_result["VID"] = data[2]
                    arp_result["Interface/Link ID"] = data[3]
                    arp_result["Aging Type"] = data[4]
                    arp_result["Type"] = data[5]
                    arp_results.append(arp_result)
            elif in_mac_section:
                data = line.split()
                if len(data) >= 4 and data[0].lower() == mac_address.lower():
                    mac_result = {}
                    mac_result["MAC Address"] = data[0]
                    mac_result["VLAN ID"] = data[1]
                    mac_result["State"] = data[2]
                    mac_result["Port/NickName"] = data[3]
                    mac_result["Aging"] = data[4]
                    mac_results.append(mac_result)

    return arp_results, mac_results

file = "diag_lv-dc1-sp2_mac-issue.txt"
mac_address = input("Enter the MAC address to search for: ")
arp_results, mac_results = search_mac(file, mac_address)

print("\nResults from display arp all:\n")
if arp_results:
    print("IP address      MAC address     VID        Interface/Link ID  Aging      Type")
    for arp_result in arp_results:
        print("{:<15} {:<15} {:<15} {:<15} {:<10} {:<5}".format(arp_result['IP address'], arp_result['MAC address'], arp_result['VID'], arp_result['Interface/Link ID'], arp_result['Aging Type'], arp_result['Type']))
else:
    print("No results found.")

print("\nResults from display mac-address:\n")
if mac_results:
    print("===============display mac-address===============")
    print("MAC Address    VLAN ID          State          Port/NickName         Aging")
    for mac_result in mac_results:
        print("{:<15} {:<15} {:<15} {:<23} {:<10}".format(mac_result['MAC Address'], mac_result['VLAN ID'], mac_result['State'], mac_result['Port/NickName'], mac_result['Aging']))
else:
    print("No results found.")
```

# Querying for a specific IP from ARP & FIB table.


No need to search different tables everytime to figure out specific IP learnings , here you see all under one roof ,i mean cell block :) 

Just key in the ip address of interest when prompted for (Enter IP address: )

What this code does -

This code is a python script written to extract information about a given IP address from a file,  for example here "hp79-30-19-10-16.txt". It reads the data from the file, and then looks for lines that contain information about the specified IP address. It then formats the relevant data into ARP and FIB information and returns the output. If the IP address is not found, it returns a message saying so.


Replace the "file_path"  with .txt file , example"file path" hp79-30-19-10-16.txt" where hp79-30-19-10-16.txt is my complete display diagnostic-information data.

```python
def extract_ip_info(file_path, ip):
    try:
        with open(file_path, "r", encoding="UTF-8") as file:
            data = file.read()
    except:
        try:
            with open(file_path, "r", encoding="ANSI") as file:
                data = file.read()
        except:
            return "Encoding error: couldn't open file with utf-8 or cp1252 encoding."

    arp_header = [
        "===============display arp all===============",
        "IP address      MAC address    VID        Interface/Link ID   Aging       Type"
    ]
    arp_info = []
    fib_header = [
        "===============display fib=============== ",
        "Destination/Mask   Nexthop           Flag     OutInterface/Token       Label"
    ]
    fib_info = []
    for line in data.split("\n"):
        if line.startswith("="):
            continue
        arp_cols = line.split()
        if len(arp_cols) >= 5 and arp_cols[0] == ip:
            arp_info.append("{:<15} {:<10} {:<10} {:<20} {:<10} {:<0}".format(arp_cols[0], arp_cols[1], arp_cols[2], arp_cols[3], arp_cols[4], arp_cols[5]))
        fib_cols = line.split()
        if len(fib_cols) >= 3 and fib_cols[0] == ip+"/32":
            fib_info.append("{:<20} {:<17} {:<10} {:<20} {:<10}".format(fib_cols[0], fib_cols[1], fib_cols[2], fib_cols[3], fib_cols[4]))
    if arp_info or fib_info:
        return "\n".join(arp_header + arp_info) + "\n\n" + "\n".join(fib_header + fib_info)
    else:
        return "IP address {} not found in arp or fib".format(ip)

file_path = "diag_lv-dc1-sp2_mac-issue.txt"
ip = input("Enter IP address: ")
ip_info = extract_ip_info(file_path, ip)
print("\n" + ip_info)
```

# Every Display Outputs!

```python
import re
import codecs

def extract_display_output(command):

    pattern = re.compile(r"^\s*=+\s*" + re.escape(command) + r"\s*=+.*?(?=^\s*=+.*?=+)((?:(?!^\s*=+.*?=+).)*)", re.DOTALL | re.IGNORECASE | re.MULTILINE)

    with codecs.open("diag_lv-dc1-sp2_mac-issue.txt", "r", encoding='utf-8', errors='ignore') as file:
        data = file.read()

    result = re.search(pattern, data)

    if result:
        return '\n'.join([line.strip() for line in result.group().strip().split('\n')])
    else:
        return "Command output not found."

display_command = input("Enter the display command: ")

output = extract_display_output(display_command)

print("\n" + output)
```

# Advanced Debug Outputs!

```python
import re

def extract_debug_output(debug_command):
    with open("vrrp_vmac.txt", "r", encoding="utf-8", errors='ignore') as file:
        data = file.read()
        
    pattern = f"{debug_command}.*?(====deb|====bcm)"
    result = re.search(pattern, data, re.DOTALL | re.IGNORECASE | re.MULTILINE)
    
    if result:
        return result.group(0)
    else:
        return "No debug output found."
debug_command = input("Enter the debug command: ")
output = extract_debug_output(debug_command)
print("\n" + output)
```

# Debugging advanced level Logbuffer

Advanced level local logbuffer display stack is executed to display debugging information for a driver module on an IRF member or Standalone device, information that could be useful for ERT & L4.

For Chassis based switches like 12900E - Enter Chassis x and Slot x in the keyword prompt.

For Slot based Switches like 5510/5490/5945- In chassis prompt click enter to bypass and slot number like 1 or 2 on slot prompt.

Just replace the filename = with diagnostic filename at the target location = "xxx.txt"

```python
filename = "display_diag_new.txt"
chassis = input("Enter the chassis number (or press enter to bypass): ")
if chassis:
    slot = input("Enter the slot number: ")
    log_start = False
    with open(filename, 'r', encoding='ansi') as f:
        for line in f:
            if "local logbuffer " + chassis + " " + slot + " display" in line:
                log_start = True
            if log_start:
                print(line.strip())
                if line.startswith("===="):
                    break
    if not log_start:
        print("No logs found for " + chassis + " " + slot)
else:
    slot = input("Enter the slot number: ")
    print("\n")
    log_start = False
    with open(filename, 'r', encoding='ansi') as f:
        for line in f:
            if "local logbuffer slot " + slot + " display" in line:
                log_start = True
            if log_start:
                print(line.strip())
                if line.startswith("===="):
                    break
    if not log_start:
        print("No logs found for slot " + slot)
```
